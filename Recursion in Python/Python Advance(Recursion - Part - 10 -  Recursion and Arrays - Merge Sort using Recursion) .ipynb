```python
def partitionFunction(l1, s, e):
    pivot = l1[e]      # Choose last element as pivot
    i = s - 1         # i tracks the smaller elements boundary

    for j in range(s, e):
        if l1[j] <= pivot:
            i += 1
            l1[i], l1[j] = l1[j], l1[i]  # Swap smaller element to left

    # Place pivot in its correct position
    l1[i+1], l1[e] = l1[e], l1[i+1]
    return i + 1       # Return pivot's final index


def QuickSort(l1, s, e):
    if s >= e:
        return

    pivotIndex = partitionFunction(l1, s, e)

    QuickSort(l1, s, pivotIndex - 1)
    QuickSort(l1, pivotIndex + 1, e)


l1 = [3, 6, 7, 2, 1, 4, 5]
QuickSort(l1, 0, len(l1)-1)
print(l1)  # [1, 2, 3, 4, 5, 6, 7]
```
