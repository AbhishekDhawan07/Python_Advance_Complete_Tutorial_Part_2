```python
def Merge(l1, s, m, e):
    left = l1[s:m+1]    # Left half
    right = l1[m+1:e+1] # Right half
    
    i = 0  # index for left
    j = 0  # index for right
    k = s  # index for main list
    
    # Compare and merge
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            l1[k] = left[i]
            i += 1
        else:
            l1[k] = right[j]
            j += 1
        k += 1
    
    # Copy remaining elements
    while i < len(left):
        l1[k] = left[i]
        i += 1
        k += 1
    
    while j < len(right):
        l1[k] = right[j]
        j += 1
        k += 1

def MergeSort(l1, s, e):
    if s >= e:
        return
    
    m = s + (e - s) // 2
    MergeSort(l1, s, m)
    MergeSort(l1, m+1, e)
    Merge(l1, s, m, e)

l1 = [1, 10, 3, 16, 9, 15]
MergeSort(l1, 0, len(l1)-1)
print(l1)  # [1, 3, 9, 10, 15, 16]
```
